```python
import pandas as pd
data = pd.read_csv('/content/TA_restaurants_curated.csv', on_bad_lines='skip')
data = data.dropna(subset=['Rating'])
data = data.dropna(subset=['City'])
data = data.dropna(subset=['Reviews'])
data = data.drop(columns=['URL_TA', 'Ranking'])
data
```

```python
# Group by 'city' and calculate the average rating
grouped_data = data.groupby('City').agg(
    average_rating=('Rating', 'mean')
).reset_index()
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Ensure you have necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Initialize stopwords
stop_words = set(stopwords.words('english'))

# Function to extract keywords based on the description
def extract_keywords(description):
    # Tokenize and remove stop words
    tokens = word_tokenize(description.lower())
    filtered_words = [word for word in tokens if word.isalnum() and word not in stop_words]
    return filtered_words

# Apply the function to the 'Description' column
data['keywords'] = data['Description'].apply(extract_keywords)

# Display the results
data[['City', 'keywords']]
```

```python
data = pd.merge(grouped_data, data, on="City", how="left")
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd


# Convert to DataFrame
city_df = data

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Fit the vectorizer to the city descriptions
city_tfidf = vectorizer.fit_transform(city_df['Description'])

# User input: preferences
user_input = "I want a city with great food and rich history"

# Convert user input to a TF-IDF vector (same vectorizer used for cities)
user_tfidf = vectorizer.transform([user_input])

# Calculate cosine similarit  y between user input and city descriptions
cosine_similarities = cosine_similarity(user_tfidf, city_tfidf)

# Add the cosine similarity scores to the city dataframe
city_df['Similarity'] = cosine_similarities.flatten()

# Apply bonus if user mentions 'food' and city's average rating > 4
food_bonus = 0.1  # A bonus value to add to the similarity score

# Check if the user mentioned 'food' in their input
if 'food' in user_input.lower():
    # Add bonus to similarity for cities with average rating > 4
    city_df.loc[city_df['average_rating'] > 4, 'Similarity'] += food_bonus

# Get the city with the highest similarity score
highest_similarity_city = city_df.loc[city_df['Similarity'].idxmax()]

# Display the city with the highest similarity
print(highest_similarity_city[['City', 'Similarity', 'average_rating']])
```
